### String formatting (3 methods)
Percent-style (%) string formatting (old style)
```
print("The answer is %6.4f +/- %8.3f" % (y, dy) )
```
str.format() (new style)
```
print("The answer is {:6.4f} +/- {:8.3f}".format(y, dy))
```
f-strings
```
print(f"The answer is {y:6.4f} +/- {dy:8.3f}")
```

### From Self-assessment question 2.3 
What will the following output?

````python
y = 9.47
dy = 0.58293
print(f"The answer is {y:6.4f} +/- {dy:8.3f}")
````

The `n` in `{dy:n.3f}` is the *minimum field width*
* Specifies minimum number of characters to display a value. 
* If actual value has fewer the field will be padded with spaces to meet minimum width. 
* If actual value has more characters, minimum field width is ignored and the full value is displayed.
* It does not have to be specified, e.g., `{dy:.3f}`

**Polling question 2.1:** Which choice will give the following output with two spaces?<br><br>
`tau:  6.28`

````python
tau = 6.283185307179586

print(f"tau: {tau:4.2f}") # A
print(f"tau: {tau:5.2f}") # B
print(f"tau: {tau:6.2f}") # C
print(f"tau:{tau:6.2f}")  # D
# More than one of these  # E
````

Why would you ever need this space padding? A useful example from https://www.pythonmorsels.com/string-formatting/

**Example:** a list of numbers formatted to be space-padded to 2 characters

````python
tracks = [(1, "Harlem"), (3, "Grandma's Hands"), (10, "Moanin' and Groanin'")]
for n, title in tracks:
    print(f"{n:2d}. {title}") # 'd' is for integer
````

````python
tracks = [(1, "Harlem"), (3, "Grandma's Hands"), (10, "Moanin' and Groanin'")]
for n, title in tracks:
    print(f"{n:02d}. {title}")
````

**Q:** What's the difference between `{tau:.3f}` and `{tau:.3}`?

### String slicing (extracting a portion of a string)

**Polling question 2.2:** How do we print `y d`?

````python
my_string = "Rainy days"

print(my_string[3:5])  # A
print(my_string[3:6])  # B
print(my_string[4:6])  # C
print(my_string[4:7])  # D
# None of these        # E
````

### String indexing (extracting a single character from a string)
Indexing: extracting a single character from a string

````python
my_string = "Rainy days"
print(my_string[2])
print(my_string[2:3]) # Same as the one above
print(my_string[-2])
````
